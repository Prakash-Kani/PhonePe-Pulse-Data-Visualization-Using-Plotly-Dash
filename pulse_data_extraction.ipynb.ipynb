```python
path_india = "D:/vscode/pulse/data/aggregated/transaction/country/india/state/"
lst=[]
aggregated_transaction_india = os.listdir(path_india)
for state in aggregated_transaction_india:
    path_state = path_india + state + '/'
    aggregated_transaction_states = os.listdir(path_state)
    for year in aggregated_transaction_states:
        path_state_year = path_state + year + '/'
        aggregated_transaction_state_year = os.listdir(path_state_year)
        for quarter in aggregated_transaction_state_year:
            path_state_year_file = path_state_year + quarter
            data = open(path_state_year_file, 'r')
            d = json.load(data)
            for l in d['data']['transactionData']:
                # n=state.replace('-', ' ')
                # m=n.title()
                # # print(m)
                data = dict(State = state,
                            Year = year,
                            Quarter = int(quarter.strip('.json')),
                            Transaction_Type = l['name'],
                            Transaction_Count =  l['paymentInstruments'][0]['count'],
                            Transaction_Amount = l['paymentInstruments'][0]['amount'])
                lst.append(data)
Agg_tran_state=pd.DataFrame(lst)
Agg_tran_state
```

```python
import os
from dotenv import load_dotenv

# Load environment variables from .env file
load_dotenv()

# MYSQL Connection
mysql_host_name = os.getenv("MYSQL_HOST_NAME")
mysql_user_name = os.getenv("MYSQL_USER_NAME")
mysql_password = os.getenv("MYSQL_PASSWORD")
mysql_database_name = os.getenv("MYSQL_DATABASE_NAME")
```

```python

from sqlalchemy import create_engine
engine = create_engine(f"mysql+mysqlconnector://{mysql_user_name}:{mysql_password}@{mysql_host_name}/{mysql_database_name}")
curs = engine.connect()
```

```python
# Transaction Table

Agg_tran_state.to_sql('Aggregated_transaction', if_exists = 'replace', index = False, con = engine)
map_tran_state.to_sql('Map_transaction', if_exists = 'replace', index = False, con = engine)
top_tran_state.to_sql('Top_transaction_district', if_exists = 'replace', index = False, con = engine)
top_tran_pincode.to_sql('Top_transaction_pincode', if_exists = 'replace', index = False, con = engine)
total_top_tran.to_sql('Top_transaction_state', if_exists = 'replace', index = False, con = engine)
```

```python
# User Table
Agg_user_state.to_sql('Aggregated_user', if_exists = 'replace', index = False, con = engine)
map_user_state.to_sql('Map_user', if_exists = 'replace', index = False, con = engine)
top_user_state_pin.to_sql('Top_user_pincode', if_exists = 'replace', index = False, con = engine)
top_user_state_dis.to_sql('Top_user_district', if_exists = 'replace', index = False, con = engine)
total_user_state.to_sql('Top_user_state', if_exists = 'replace', index = False, con = engine)
```
